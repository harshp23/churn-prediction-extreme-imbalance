# Revenue and orders Trend (monthly)

```python
query = '''select 
 date_format(o.order_purchase_timestamp , '%Y-%m-01') As month_start,
 count(distinct o.order_id) as orders_count,
 sum(p.payment_value) as gross_revenue,
 sum(p.payment_value) / nullif(count(distinct o.order_id),0) as avg_order_value
from orders_cleaned o
join payments p on p.order_id =  o.order_id
where o.order_status = 'delivered'
group by 1
order by 1;'''

cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data,columns=['month_start','order_count','gross_revenue','avg_order_value'])
df['month_start'] = pd.to_datetime(df['month_start'])

# plot Revenue over time
plt.figure(figsize=(10,5))
plt.plot(df['month_start'],df['gross_revenue'])
plt.title("Monthly Gross Revenue (Delivered Orders)")
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation= 45)
plt.tight_layout()
plt.show()
```

```python
# orders over time

plt.figure(figsize=(10,5))
plt.plot(df['month_start'],df['order_count'])
plt.title('Monthly orders count ( Delivered Orders)')
plt.xlabel('Month')
plt.ylabel('Orders')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Plot : Average Order Value(AOV)
plt.figure(figsize = (10,5))
plt.plot(df['month_start'],df['avg_order_value'])
plt.title('Monthly Average Order value ')
plt.xlabel('Month')
plt.ylabel('AOV')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

# Delivery time Distribution


```python
query = ''' SELECT 
    DATEDIFF(o.order_delivered_customer_date, o.order_purchase_timestamp) AS delivery_days,
    COUNT(*) AS num_orders
FROM orders_cleaned o
WHERE o.order_status = 'delivered'
GROUP BY delivery_days
ORDER BY delivery_days;
'''
cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data,columns=['Delivery_days','num_orders'])

sns.histplot(df['Delivery_days'],bins=30,kde=True)
plt.title("Delivery Time Distribution (Days)")
plt.xlabel("Days")
plt.ylabel('Number of Orders')
plt.show()
```

# Late vs On-time deliveries

```python
query = '''SELECT 
    CASE 
        WHEN o.order_delivered_customer_date <= o.order_estimated_delivery_date THEN 'On Time'
        ELSE 'Late'
    END AS delivery_status,
    COUNT(*) AS num_orders
FROM orders_cleaned o
WHERE o.order_status = 'delivered'
GROUP BY delivery_status;
'''
cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data , columns=['Delivery_status','num_orders'])

sns.barplot(x= df['Delivery_status'],y=df['num_orders'],data = df)
plt.show()
# df.value_counts()
```

# Delivery Delay Hotspots by state/city

```python
query = '''SELECT 
    c.customer_state,
    AVG(DATEDIFF(o.order_delivered_customer_date, o.order_estimated_delivery_date)) AS avg_delay_days,
    COUNT(*) AS num_orders
FROM orders_cleaned o
JOIN customers c ON o.customer_id = c.customer_id
WHERE o.order_status = 'delivered'
GROUP BY c.customer_state
ORDER BY avg_delay_days DESC;'''

cur.execute(query)
data = cur.fetchall()
df = pd.DataFrame(data,columns=['customer_state','avg_delay_days','num_orders',])

plt.figure(figsize=(10,5))
sns.barplot(x=df['customer_state'],y=df['avg_delay_days'])
plt.show()
```

# Customer and churn analysis

```python
# purchase frequency per customer
query = '''SELECT 
    c.customer_unique_id, 
    COUNT(DISTINCT o.order_id) AS purchase_count
FROM orders_cleaned o join customers c on c.customer_id = o.customer_id
GROUP BY customer_unique_id
order by purchase_count desc;
 '''

cur.execute(query)
data = cur.fetchall()
df = pd.read_sql(query,db)

# one-time vs repeat customers
one_time = (df['purchase_count'] ==1).sum()
repeat = (df['purchase_count']>1).sum()

print(f"one time customers : {one_time}")
print(f'Repeat customers : {repeat}')

# Distribution of purchase frequency
freq_distribution = df['purchase_count'].value_counts().sort_index()

plt.figure(figsize=(8,5))
freq_distribution.plot(kind="bar")
plt.xlabel("Number of Orders per Customer")
plt.ylabel("Number of Customers")
plt.title("Customer Purchase Frequency Distribution")
plt.show()

# 4. Percentages
total_customers = len(df)
print("One-time Customers %:", round(one_time/total_customers*100, 2))
print("Repeat Customers %:", round(repeat/total_customers*100, 2))


```
