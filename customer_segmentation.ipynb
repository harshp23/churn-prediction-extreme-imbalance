```python
import pandas as pd

# Load datasets
orders = pd.read_csv(r'C:\Users\harsh\OneDrive\Desktop\data analysis\Agentic AI-Powered Revenue & Churn Analytics on E-Commerce Data\data\clean_orders.csv')
order_items = pd.read_csv(r"C:\Users\harsh\OneDrive\Desktop\data analysis\Agentic AI-Powered Revenue & Churn Analytics on E-Commerce Data\data\olist_order_items_dataset.csv")
customers = pd.read_csv(r"C:\Users\harsh\OneDrive\Desktop\data analysis\Agentic AI-Powered Revenue & Churn Analytics on E-Commerce Data\data\olist_customers_dataset.csv")

# Convert to datetime
orders['order_purchase_timestamp'] = pd.to_datetime(orders['order_purchase_timestamp'])

# Merge customers with orders
orders_customers = pd.merge(orders, customers, on="customer_id", how="left")

# Merge order_items to get monetary values
orders_items = pd.merge(orders_customers, order_items, on="order_id", how="left")

# Monetary: total spend per order
orders_items['payment_value'] = orders_items['price'] + orders_items['freight_value']

# Aggregate at customer_unique_id level
rfm = orders_items.groupby("customer_unique_id").agg({
    "order_purchase_timestamp": lambda x: (orders['order_purchase_timestamp'].max() - x.max()).days,  # Recency
    "order_id": "nunique",  # Frequency = number of unique orders
    "payment_value": "sum"  # Monetary = total spend
}).reset_index()

# Rename columns
rfm.rename(columns={
    "order_purchase_timestamp": "Recency",
    "order_id": "Frequency",
    "payment_value": "Monetary"
}, inplace=True)

# Quick look
print(rfm.head())

# Optional: RFM Scoring (1â€“5 scale)
rfm['R_Score'] = pd.qcut(rfm['Recency'], 5, labels=[5,4,3,2,1])
rfm['F_Score'] = pd.qcut(rfm['Frequency'].rank(method="first"), 5, labels=[1,2,3,4,5])
rfm['M_Score'] = pd.qcut(rfm['Monetary'], 5, labels=[1,2,3,4,5])

rfm['RFM_Segment'] = rfm['R_Score'].astype(str) + rfm['F_Score'].astype(str) + rfm['M_Score'].astype(str)
rfm['RFM_Score'] = rfm[['R_Score','F_Score','M_Score']].astype(int).sum(axis=1)

print(rfm.head())
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# Recency
axes[0].hist(rfm['Recency'], bins=30, edgecolor='black')
axes[0].set_title('Recency Distribution')
axes[0].set_xlabel('Days since last purchase')
axes[0].set_ylabel('Number of Customers')

# Frequency
axes[1].hist(rfm['Frequency'], bins=30, edgecolor='black')
axes[1].set_title('Frequency Distribution')
axes[1].set_xlabel('Number of Orders')
axes[1].set_ylabel('Number of Customers')

# Monetary
axes[2].hist(rfm['Monetary'], bins=30, edgecolor='black')
axes[2].set_title('Monetary Distribution')
axes[2].set_xlabel('Total Spend (BRL)')
axes[2].set_ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```

# Churn labeling

```python
import pandas as pd

# churn threshold
churn_threshold = 180

# create churn label
rfm['churn_label'] = rfm['Recency'].apply(lambda X: 1 if X> churn_threshold else 0)

print(rfm['churn_label'].value_counts())
```

```python
rfm.to_csv("rfm_with_churn_labels.csv",index=False)
print("Churn labelling done. File saved as rfm_with_churn_labels.csv")
```
