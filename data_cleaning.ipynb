```python
import pandas as pd
import numpy as np
```

```python
# load dataset
orders = pd.read_csv('olist_orders_dataset.csv')
customers = pd.read_csv('olist_customers_dataset.csv')
items = pd.read_csv('olist_order_items_dataset.csv')
products = pd.read_csv('olist_products_dataset.csv')
sellers = pd.read_csv('olist_sellers_dataset.csv')
reviews = pd.read_csv('olist_order_reviews_dataset.csv')
payments = pd.read_csv('olist_order_payments_dataset.csv')
category_transaction = pd.read_csv('product_category_name_translation.csv')
```

```python
# Preview shapes
print("Orders:", orders.shape)
print("Customers:", customers.shape)
print("Items:", items.shape)
print("Products:", products.shape)
print("Sellers:", sellers.shape)
print("Reviews:", reviews.shape)
print("Payments:", payments.shape)
```

```python
# checking missing values
print("\nMissing Values:")
for name, df in zip(
    ["Orders","Customers","Items","Products","Sellers","Reviews","Payments"],
    [orders, customers, items, products, sellers, reviews, payments]
):
    print(f"{name}:")
    print(df.isnull().sum())
    print("-"*40)
```

```python
# check duplicates
print('\n Duplicates rows check :')
print('Orders :',orders.duplicated().sum())
print('customers :',customers.duplicated().sum())
print('Items :',items.duplicated().sum())
```

```python
# filling No comments to title and message columns 
reviews['review_comment_title'] = reviews['review_comment_title'].fillna("No Comments")
reviews['review_comment_message'] = reviews['review_comment_message'].fillna('No Comments')
```

```python
# products Cleaning 
products['product_category_name'] = products['product_category_name'].fillna("unknown_category")
```

```python
# for numarical missing values imputation we check distriution of points
import matplotlib.pyplot as plt
import seaborn as sns

# numerical columns in produts with missing values
num_col =['product_weight_g',                
'product_length_cm',             
'product_height_cm',               
'product_width_cm' ]

# plot distribution and boxplots
for col in num_col:
    plt.figure(figsize=(12,5))

    # histogram + kde
    plt.subplot(1,2,1)
    sns.histplot(products[col],bins=50,kde=True)
    plt.title(f"Distribution of {col}")

    # boxplot
    plt.subplot(1,2,2)
    sns.boxplot(x=products[col])
    plt.title(f"Boxplot of {col}")
    plt.show()
```

```python
# All Above columnns are right skewed and having outliers so we fill missing values with median
num_col =['product_weight_g',                
'product_length_cm',             
'product_height_cm',               
'product_width_cm',
'product_name_lenght','product_description_lenght','product_photos_qty' ]

for col in num_col:
    products[col] = products[col].fillna(products[col].median())
```

```python
orders.isnull().sum()
```

```python
# Fill order_approved at with purchase time if missing
orders['order_approved_at'].fillna(orders['order_purchase_timestamp'],inplace=True)

# Add buisness flag
orders['is_shipped'] =orders['order_delivered_carrier_date'].notna()
orders['is_delivered'] = orders['order_delivered_customer_date'].notna()
```

```python
# Save cleaned versions
orders.to_csv("clean_orders.csv", index=False)
products.to_csv("clean_products.csv", index=False)
reviews.to_csv("clean_reviews.csv", index=False)
```
